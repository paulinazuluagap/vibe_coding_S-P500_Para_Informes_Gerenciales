###**Paso 1:** Fotografía del Mercado 📸

1.   ¿Qué sectores son los más grandes en términos de valor de mercado?
2.   ¿Qué empresas individuales son las más dominantes?


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de los gráficos
sns.set(style="whitegrid")

# --- CÁLCULO ---
# Agrupamos por 'Sector' y sumamos la 'Marketcap'
sector_marketcap = df_companies.groupby('Sector')['Marketcap'].sum().sort_values(ascending=False)

# Convertimos a billones para que sea más fácil de leer
sector_marketcap_billions = sector_marketcap / 1e9

# --- VISUALIZACIÓN ---
plt.figure(figsize=(12, 8)) # Tamaño del gráfico
ax = sns.barplot(x=sector_marketcap_billions.values, y=sector_marketcap_billions.index, palette='viridis')

# Añadir títulos y etiquetas
plt.title('Dominancia de Mercado por Sector en el S&P 500', fontsize=16, weight='bold')
plt.xlabel('Capitalización de Mercado Total (en Billones de USD)', fontsize=12)
plt.ylabel('Sector', fontsize=12)

# Añadir el valor exacto en cada barra para mayor claridad
for i in ax.containers:
    ax.bar_label(i, fmt='B $%.0f', label_type='edge', padding=5, fontsize=10)

plt.show()

# --- TABLA DE DATOS ---
print("\nTabla de Capitalización por Sector (en Billones de USD):")
print(sector_marketcap_billions.to_frame(name="Marketcap (Billions)"))
```

##**Explicación:**
El análisis de la composición del S&P 500 muestra una dominancia abrumadora del sector de Tecnología, que posee una capitalización de mercado total de 17.3 billones de dólares (17,287 trillions). La magnitud de este liderazgo es tal, que su valor es más del doble que el de su seguidor más cercano, Servicios de Comunicación, que se sitúa en $7.8 trillones.

Un segundo escalón de influencia está compuesto por los sectores de Servicios Financieros (7.1 billones de dólares), Consumo Cíclico (6.4 billones de dólares) y Salud (5.2 billones de dólares). Estos tres, junto con Servicios de Comunicación, forman un bloque robusto que, sin embargo, opera a una escala significativamente menor que la del sector tecnológico.

El resto de los sectores, como Industriales, Energía y Materiales Básicos, tienen una ponderación considerablemente inferior, lo que demuestra que el valor del mercado accionario actual está fuertemente concentrado en la economía digital y los servicios.

###**Paso 3:** Análisis de Rendimiento Histórico 📈

Ahora vamos a crear el gráfico comparativo de rendimiento. El objetivo es ver visualmente cómo se han comportado los sectores más grandes en comparación con el mercado en general (el índice S&P 500).

```python
# ===================================================================
# PASO 2: CÓDIGO COMPLETO Y CORREGIDO FINAL (EJECUTAR EN UNA SOLA CELDA)
# ===================================================================
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# --- Parte 1: Preparación de los Datos ---
df_index['Date'] = pd.to_datetime(df_index['Date'])
df_stocks['Date'] = pd.to_datetime(df_stocks['Date'])
df_stocks = df_stocks.drop_duplicates(subset=['Date', 'Symbol'])

# << AQUÍ ESTÁ LA CORRECCIÓN: Se cambió 'left' por how='left' >>
stocks_with_sector = pd.merge(df_stocks, df_companies[['Symbol', 'Sector']], on='Symbol', how='left')
stocks_with_sector.dropna(subset=['Sector'], inplace=True)

# --- Parte 2: Cálculo de Rendimientos Acumulados ---
# A. Para el Índice S&P 500
df_index_copy = df_index.set_index('Date').copy()
index_daily_returns = df_index_copy['S&P500'].pct_change().dropna()
index_cumulative_returns = (1 + index_daily_returns).cumprod()

# B. Para cada Sector
stocks_pivot = stocks_with_sector.pivot_table(index='Date', columns='Symbol', values='Close')
daily_returns_stocks = stocks_pivot.pct_change()
symbol_to_sector = df_companies.set_index('Symbol')['Sector'].to_dict()
sector_daily_returns = daily_returns_stocks.rename(columns=symbol_to_sector).T.groupby(level=0).mean().T
sector_cumulative_returns = (1 + sector_daily_returns).cumprod()

# --- Parte 3: Visualización Unificada ---
if 'sector_marketcap' not in locals():
    sector_marketcap = df_companies.groupby('Sector')['Marketcap'].sum().sort_values(ascending=False)
top_5_sectors = sector_marketcap.head(5).index.tolist()

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(15, 8))

# Graficar el S&P 500 (línea base)
(index_cumulative_returns).plot(label='S&P 500 Index', color='black', linewidth=3, linestyle='--')

# Graficar los 5 sectores principales en el MISMO gráfico
(sector_cumulative_returns[top_5_sectors]).plot(linewidth=2)

# Añadir títulos y etiquetas
plt.title('Rendimiento Comparativo: Sectores vs. S&P 500', fontsize=18, weight='bold')
plt.ylabel('Crecimiento de $1 invertido', fontsize=12)
plt.xlabel('Fecha', fontsize=12)
plt.legend(title='Leyenda', fontsize='large')
plt.grid(True)
plt.show()
```

##**Explicación:**
El análisis de rendimiento histórico demuestra que, si bien el mercado en su conjunto ha tenido un crecimiento robusto, ciertos sectores han sido verdaderos motores de creación de valor, superando ampliamente el rendimiento promedio del S&P 500.

**El sector de Tecnología** emerge como el líder indiscutible. Durante el período analizado, una inversión en este sector se habría multiplicado casi por 18, un rendimiento extraordinario que eclipsa a todos los demás. Le siguen, con un desempeño también sobresaliente, los sectores de **Consumo Cíclico y Servicios de Comunicación**, que multiplicaron la inversión inicial aproximadamente por 15 y 11 veces, respectivamente.

Es crucial notar que los cinco sectores más grandes del índice lograron superar el rendimiento del S&P 500. Mientras que el índice general (visible en el gráfico de la línea negra) convirtió 1 en aproximadamente 3, incluso el sector con el rendimiento más bajo de este grupo selecto, **Servicios Financieros**, multiplicó la inversión por más de 7. Esto indica que el crecimiento del mercado ha sido impulsado de manera desproporcionada por sus componentes más grandes y establecidos.

###**Paso 4:** Matriz de Riesgo Vs. Rendimiento 🎯

Ya sabemos qué sectores fueron más rentables, pero ¿fueron también los más arriesgados? Un alto rendimiento acompañado de una volatilidad extrema puede no ser ideal.

En este paso, crearemos un gráfico de dispersión (scatter plot) para visualizar cada sector en una matriz. Esto nos permitirá clasificar los sectores en cuadrantes y encontrar el "punto ideal" (alto rendimiento con bajo riesgo).

**Eje Y (Rendimiento):** Rentabilidad promedio anual.

**Eje X (Riesgo):** Volatilidad (desviación estándar) anual.

```python
import numpy as np

# --- Parte 1: Cálculo de Métricas de Riesgo y Rendimiento ---

# Usaremos los retornos diarios por sector que ya calculamos
# Hay aproximadamente 252 días de mercado en un año

# 1. Calcular el rendimiento promedio anualizado para cada sector
annual_return = sector_daily_returns.mean() * 252

# 2. Calcular la volatilidad (riesgo) anualizada para cada sector
annual_volatility = sector_daily_returns.std() * np.sqrt(252)

# 3. Crear un nuevo DataFrame con estos resultados
risk_return_df = pd.DataFrame({'Rendimiento_Anual': annual_return, 'Volatilidad_Anual': annual_volatility})


# --- Parte 2: Visualización de la Matriz ---

plt.figure(figsize=(12, 8))
sns.scatterplot(data=risk_return_df, x='Volatilidad_Anual', y='Rendimiento_Anual', s=100) # s es el tamaño de los puntos

# Añadir etiquetas a cada punto para saber qué sector es
for i in range(risk_return_df.shape[0]):
    plt.text(x=risk_return_df.Volatilidad_Anual[i]+0.005, # Pequeño ajuste para que no se solape
             y=risk_return_df.Rendimiento_Anual[i],
             s=risk_return_df.index[i], # El nombre del sector
             fontdict=dict(color='black', size=10))

# Añadir líneas de los promedios para crear los 4 cuadrantes
plt.axvline(risk_return_df.Volatilidad_Anual.mean(), color='grey', linestyle='--', label='Volatilidad Promedio')
plt.axhline(risk_return_df.Rendimiento_Anual.mean(), color='red', linestyle='--', label='Rendimiento Promedio')

# Títulos y etiquetas
plt.title('Matriz de Riesgo vs. Rendimiento por Sector', fontsize=18, weight='bold')
plt.xlabel('Riesgo (Volatilidad Anualizada)', fontsize=12)
plt.ylabel('Rendimiento (Promedio Anualizado)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# Opcional: Mostrar la tabla con los datos calculados
print("\n--- Datos de Riesgo y Rendimiento Anualizado por Sector ---")
display(risk_return_df.style.format({
    'Rendimiento_Anual': '{:.2%}',
    'Volatilidad_Anual': '{:.2%}'
}))
```

##**Explicación:**
La matriz de riesgo vs. rendimiento nos permite evaluar la eficiencia de cada sector, analizando qué tanta rentabilidad han generado en proporción a la volatilidad que han experimentado. El análisis posiciona a los sectores en cuatro perfiles estratégicos claros:

**1. El Punto Ideal: Alto Rendimiento con Bajo Riesgo**  
En el cuadrante superior izquierdo se encuentran los sectores más eficientes. Healthcare (Salud) y Communication Services (Servicios de Comunicación) destacan en esta categoría, habiendo entregado rendimientos superiores al promedio con una volatilidad inferior. Estos sectores han demostrado un balance óptimo, ofreciendo un crecimiento sólido sin exponer a los inversores a un riesgo excesivo.

**2. Líderes de Alto Potencial: Alto Rendimiento con Alto Riesgo**  
El cuadrante superior derecho está ocupado por los motores de crecimiento del mercado. Technology (Tecnología) es el líder indiscutible en rendimiento, pero también presenta una de las volatilidades más altas. Junto a Consumer Cyclical (Consumo Cíclico), estos sectores han sido los más rentables, pero este desempeño ha venido acompañado de un riesgo considerable, perfilándolos como opciones para estrategias de inversión más agresivas.

**3. El Refugio Conservador: Bajo Riesgo con Bajo Rendimiento**  
El cuadrante inferior izquierdo es el dominio de los sectores defensivos. Consumer Defensive (Consumo Defensivo) y, especialmente, Utilities (Servicios Públicos), se caracterizan por su baja volatilidad. Aunque su rentabilidad es modesta, su principal atractivo es la estabilidad y la previsibilidad, actuando como un ancla en tiempos de incertidumbre.

**4. El Dilema de la Ineficiencia: Bajo Rendimiento con Alto Riesgo**  
Finalmente, el cuadrante inferior derecho revela a los sectores menos eficientes. Energy (Energía) y Basic Materials (Materiales Básicos) presentan el peor perfil, habiendo combinado una alta volatilidad con rendimientos por debajo del promedio. Financial Services (Servicios Financieros) se encuentra en una posición de dilema, ya que su rendimiento apenas alcanza el promedio, pero lo hace con un nivel de riesgo superior a la media del mercado.
